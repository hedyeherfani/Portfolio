```python
import pandas as pd 
```

```python
sheets = pd.read_excel('online_retail.xlsx',sheet_name = ['Year 2009-2010','Year 2010-2011'])
```

```python
dataset = pd.concat(sheets[frame] for frame in sheets.keys())
```

```python
dataset.reset_index(drop=True)
```
