```python
import pandas as pd
```

```python
dataset = pd.read_csv('step2',low_memory = False)
```

```python
dataset.head()
```

```python
del dataset['Unnamed: 0']
```

```python
dataset.head()
```

```python
 dataset.dropna(subset=['Customer ID'])
```

```python
dataset['Customer ID'] = dataset['Customer ID'].astype(str)
```

```python
dataset.head()
```

```python
dataset['InvoiceDate'] = pd.to_datetime(dataset['InvoiceDate'])
```

```python
dataset.dtypes
```

```python
today = pd.datetime(2012,1,1)
```

```python
days_since_today = today-dataset.InvoiceDate
```

```python
dataset['days_since_today'] = days_since_today
```

```python
dataset['days_since_today'] = days_since_today.dt.components.days 
```

```python
dataset.head()
```

```python
days_since_purchased = dataset.groupby('Customer ID').agg({'days_since_today': ['min']})
```

```python
days_since_purchased.drop('nan')
```

```python
Recency = days_since_purchased
Recency
```

```python
Frequency = dataset.drop_duplicates(subset = "Invoice")['Customer ID'].value_counts()
```

```python
Frequency = Frequency.to_frame()
```

```python
Frequency = Frequency.iloc[1:]
Frequency
```
